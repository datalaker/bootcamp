## Dropbox

### Salaries Differences

Write a query that calculates the difference between the highest salaries found in the marketing and engineering departments. Output just the absolute difference in salaries.

```python
db_employee = read_csv('employees')
db_employee.head()
```

```python
db_dept = read_csv('dept')
db_dept.head()
```

```python
df = pd.merge(db_employee, db_dept, how = 'left',left_on = ['department_id'], right_on=['id'])

df1 = df[df["department"]=='engineering']
df_eng = df1.groupby('department')['salary'].max().reset_index(name='eng_salary')

df2 = df[df["department"]=='marketing']
df_mkt = df2.groupby('department')['salary'].max().reset_index(name='mkt_salary')

result = abs(pd.DataFrame(df_mkt['mkt_salary'] - df_eng['eng_salary']))
result.columns = ['salary_difference']
result
```

### Bikes Last Used

Find the last time each bike was in use. Output both the bike number and the date-timestamp of the bike's last use (i.e., the date-time the bike was returned). Order the results by bikes that were most recently used.

## Yelp

### Top Businesses With Most Reviews

Find the top 5 businesses with most reviews. Assume that each row has a unique business_id such that the total reviews for each business is listed on each row. Output the business name along with the total number of reviews and order your results by the total reviews in descending order.

```python
yelp_business = pd.read_csv(os.path.join(ROOT_DATA_PATH, 'yelp_business.csv'), sep='|')
yelp_business
```

```python
yelp_business['rank'] = yelp_business['review_count'].rank(ascending=False)
result = yelp_business[yelp_business['rank']<=5][['name', 'review_count']].sort_values(by='review_count', ascending=False)
result.head()
```

## Amazon

### Customer Details

Find the details of each customer regardless of whether the customer made an order. Output the customer's first name, last name, and the city along with the order details.
You may have duplicate rows in your results due to a customer ordering several of the same items. Sort records based on the customer's first name and the order details in ascending order.

```python
amazon_customers = read_csv('amazon_customers')
amazon_customers.head()
```

```python
amazon_orders = read_csv('amazon_orders')
amazon_orders.head()
```

```python
merged = pd.merge(amazon_customers, amazon_orders, left_on = 'id', right_on = 'cust_id', how = 'left')
result = merged[['first_name','last_name','city','order_details']].sort_values(['first_name','order_details'])
result.head()
```

### Order Details

Find order details made by Jill and Eva.
Consider the Jill and Eva as first names of customers.
Output the order date, details and cost along with the first name.
Order records based on the customer id in ascending order.

## Apple

### MacBookPro users events

Count the number of user events performed by MacBookPro users.
Output the result along with the event name.
Sort the result based on the event count in the descending order.

```python
macbook_events = read_csv('macbook_events')
macbook_events.head()
```

```python
macbook_device = macbook_events[macbook_events['device']=='macbook pro']
result = macbook_device.groupby(['event_name']).size().to_frame('size').reset_index().sort_values(['size'], ascending= False).rename(columns={"size": "event_count"})
result.head()
```

## Salesforce

### Average Salaries

Compare each employee's salary with the average salary of the corresponding department.
Output the department, first name, and salary of employees along with the average salary of that department.

```python
salesforce_employees = read_csv('salesforce_employees')
salesforce_employees.head()
```

```python
salesforce_employees['avg_salary'] = salesforce_employees.groupby(['department'])['salary'].transform('mean')
result = salesforce_employees[['department','first_name','salary','avg_salary']]
result.head()
```
