```python
%%sql
select EXTRACT(
        MONTH
        FROM date
    ) AS MONTH,
    day_of_week,
    avg(COUNT_INC) as AVG_INC
from (
        select day_of_week,
            date,
            count(incidnt_num) as COUNT_INC
        from sf_crime_incidents_2014_01_preprocessed
        group by 1,
            2
    ) table1
group by 1,
    2
order by AVG_INC DESC
limit 10
```

```python
%%sql
select start_time,
    start_terminal,
    duration_seconds,
    sum(duration_seconds) over (
        partition by start_terminal
        order by start_time
    ) as RUNNING_TOTAL,
    avg(duration_seconds) over (
        partition by start_terminal
        order by start_time
    ) as RUNNING_TOTAL_AVG,
    count(duration_seconds) over (
        partition by start_terminal
        order by start_time
    ) as RUNNING_TOTAL_COUNT
from dc_bikeshare_q1_2012_preprocessed
where start_time < '2012-01-08'
limit 20
```

```python
%%sql
select start_time,
    start_terminal,
    duration_seconds,
    row_number() over (
        partition by start_terminal
        order by start_time 
    ) as COUNT_ROW,
    sum(duration_seconds) over (
        partition by start_terminal
        order by start_time
    ) as RUNNING_TOTAL_BY_PARTITION    
from dc_bikeshare_q1_2012_preprocessed
where start_time < '2012-01-08'
limit 20
```

```python
%%sql
select start_time,
    start_terminal,
    duration_seconds,
    row_number() over (
        partition by start_terminal
        order by start_time 
    ) as COUNT_ROW,
    sum(duration_seconds) over (
        partition by start_terminal
        order by start_time
    ) as RUNNING_TOTAL_BY_PARTITION,
    rank() over (
        partition by start_terminal
        order by start_time
    ) as RANK_PART
from dc_bikeshare_q1_2012_preprocessed
where start_time < '2012-01-08'
limit 20
```

```python
%%sql
select start_terminal,
    duration_seconds, lag(duration_seconds, 1) over (partition by start_terminal order by duration_seconds) as LAG_PART,
    duration_seconds - lag(duration_seconds, 1) over (partition by start_terminal order by duration_seconds) as LAG_DIFF
from dc_bikeshare_q1_2012_preprocessed
where start_time < '2012-01-08'
limit 20
```

```python
%%sql
select start_terminal,
    duration_seconds, lag(duration_seconds, 2) over (partition by start_terminal order by duration_seconds) as LAG_PART,
    lead(duration_seconds, 1) over (partition by start_terminal order by duration_seconds) as LEAD_PART
from dc_bikeshare_q1_2012_preprocessed
where start_time < '2012-01-08'
limit 20
```
