```python
%%sql
select permalink,
    founded_at,
    acquired_at,
    TIMESTAMPDIFF(MONTH, founded_at, acquired_at) AS GROWTH_TIME_IN_MONTHS
from cr_comp
    join cr_acq on cr_comp.permalink = cr_acq.company_permalink
having GROWTH_TIME_IN_MONTHS >= 0
order by GROWTH_TIME_IN_MONTHS
limit 10
```

```python
%%sql
select permalink,
    founded_at,
    acquired_at,
    timestampdiff(year, founded_at,now()) AS comp_age,
    DATE_ADD(acquisitions.acquired_at_cleaned, INTERVAL 1 YEAR) AS year_after_acquired
    TIMESTAMPDIFF(MONTH, founded_at, acquired_at) AS GROWTH_TIME_IN_MONTHS
from cr_comp
    join cr_acq on cr_comp.permalink = cr_acq.company_permalink
having GROWTH_TIME_IN_MONTHS >= 0
order by GROWTH_TIME_IN_MONTHS
limit 10
```

```python
%%sql
select permalink,
    founded_at,
    acquired_at,
    TIMESTAMPDIFF(year, founded_at, now()) AS COMP_AGE,
    DATE_SUB(founded_at, INTERVAL 1 YEAR) AS year_before_founded,
    DATE_ADD(founded_at, INTERVAL 1 YEAR) AS year_after_founded
from cr_comp
    join cr_acq on cr_comp.permalink = cr_acq.company_permalink
having COMP_AGE >= 0
order by COMP_AGE desc
limit 10
```

```python
%%sql
select permalink,
    left(permalink, 8) as Type,
    right(permalink, length(permalink) -9) as company_name,
    trim(both '/' from left(permalink, 8)) as trimmed_type
from cr_comp
limit 10
```

```python
%%sql
select permalink,
    left(permalink, 8) as TYPE,
    right(permalink, length(permalink) -9) as COMPANY_NAME,
    trim(LEADING '/' from left(permalink, 8)) as TRIMMED_TYPE,
    position('f' in permalink) AS POS_F_CHAR
from cr_comp
limit 10
```

```python
%%sql
select permalink,
    left(permalink, 8) as TYPE,
    right(permalink, length(permalink) -9) as COMPANY_NAME,
    trim(LEADING '/' from left(permalink, 8)) as TRIMMED_TYPE,
    position('a' in permalink) AS POS_F_CHAR
from cr_comp
limit 10
```
