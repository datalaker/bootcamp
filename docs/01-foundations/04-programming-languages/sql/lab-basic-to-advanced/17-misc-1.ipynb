```python
%%sql
select comp.permalink, comp.founded_at,
acq.acq_date 
from comp join acq 
on comp.permalink=acq.company_permalink
```

```python
%%sql
select comp.permalink, comp.founded_at,
acq.acq_date , 
TIMESTAMPDIFF(MONTH, comp.founded_at, acq.acq_date ) AS diff
from comp join acq 
on comp.permalink=acq.company_permalink
having diff >=0
Order by diff 
limit 10
```

```python
%%sql
select * from (
select comp.permalink, comp.founded_at,
acq.acq_date , 
TIMESTAMPDIFF(MONTH , comp.founded_at, now() ) AS diff
from comp join acq 
on comp.permalink=acq.company_permalink
Order by diff ) acq2 
where diff >=0
limit 10
```

```python
%%sql
SELECT comp.permalink,
    comp.founded_at,
    acq.acq_date,
    DATEDIFF(acq.acq_date, comp.founded_at) AS time_to_acquisition,
    DATE_SUB(acq.acq_date, INTERVAL 1 YEAR) AS year_before_acquired,
    DATE_ADD(acq.acq_date, INTERVAL 1 YEAR) AS year_after_acquired
FROM comp
    JOIN acq
        ON acq.company_permalink = comp.permalink
WHERE founded_at IS NOT NULL
LIMIT 10
```

```python
%%sql
SELECT founded_at,
       EXTRACT(YEAR FROM founded_at) AS year,
       EXTRACT(MONTH  FROM founded_at) AS month,
       EXTRACT(DAY  FROM founded_at) AS day,
       EXTRACT(HOUR  FROM founded_at) AS hour,
       EXTRACT(MINUTE  FROM founded_at) AS minute,
       EXTRACT(SECOND  FROM founded_at) AS second
       from comp limit 5
```
