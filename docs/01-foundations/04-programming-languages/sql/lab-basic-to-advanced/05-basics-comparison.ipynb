```python
%%sql

SELECT *
FROM us_housing_units
WHERE month_name != 'January'
LIMIT 10;
```

```python
%%sql

SELECT *
FROM us_housing_units
WHERE month_name > '"January"'
LIMIT 10;
```

```python
%%sql

SELECT *
FROM us_housing_units
WHERE month_name > '"M"'
LIMIT 10;
```

```python
%%sql

SELECT year,
       month,
       west,
       south,
       west + south AS south_plus_west
FROM us_housing_units
LIMIT 10;
```

```python
%%sql

SELECT year,
       month,
       west,
       south,
       (west + south)/2 AS south_west_avg
FROM us_housing_units
LIMIT 10;
```
