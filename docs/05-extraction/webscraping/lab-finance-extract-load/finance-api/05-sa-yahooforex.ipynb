```python
def yahoo_forex_extract(symbol):
    data = yf.Ticker(symbol)
    data = data.history(period='max')
    data.reset_index(inplace=True)
    data['Date'] = pd.to_datetime(data['Date'])
    mask = (data['Date'] > '2014-01-01') & (data['Date'] <= '2022-02-18')
    data = data.loc[mask]
    data.set_index('Date', inplace=True)
    return data
```

```python
data = [yahoo_forex_extract(symbol) for symbol in symbols]
_data = {k:v for k,v in zip(symbols, data)}
df = pd.concat(_data, axis=1)
df
```
