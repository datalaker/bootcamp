```python
import numpy as np
import pandas as pd
import random
import time
from datetime import datetime
import warnings
warnings.filterwarnings('ignore')
from faker import Faker
faker = Faker()
from faker.providers import DynamicProvider
```

```python
def ccas_cat(n, categories, ratios=None):
    if ratios is None:
        ratios = [1/len(categories)]*len(categories)
    return (np.random.choice(categories, size=(n,), p=[float(i)/sum(ratios) for i in ratios])).tolist()

def ccas_cont(n, dist, a, b, maxn=1, minn=0):
    if dist=='norm':
        return [int(i) for i in (np.random.normal(a,b,n))]
    elif dist=='beta':
        return [int(i) for i in (minn + (np.random.beta(a, b, n) * (maxn - minn)))]
    elif dist=='weibull':
        return [int(i) for i in (minn + (np.random.weibull(a, n) * (maxn - minn)))]

def randomDate(start, end):
    frmt = '%d-%m-%Y %H:%M'
    stime = time.mktime(time.strptime(start, frmt))
    etime = time.mktime(time.strptime(end, frmt))
    ptime = stime + random.random() * (etime - stime)
    dt = datetime.fromtimestamp(time.mktime(time.localtime(ptime)))
    return dt
```

```python
seed = 8203
np.random.seed(seed)
faker.seed_instance(seed)
```

```python
p = 5
df = pd.DataFrame()
df['wallet_id'] = [faker.ssn() for i in range(p)]
df['type'] = ccas_cat(p, ['giftcard','cancellation','goodwill'], ratios=[40,40,20])
df['start_date'] = [faker.date_time_between_dates(datetime_start=datetime(2021,1,1),
                                                  datetime_end=datetime(2022,3,30))\
                         .strftime("%Y/%m/%d, %H:%M:%S") for i in range(p)]
df['expiry_date'] = [faker.date_time_between_dates(datetime_start=datetime(2022,4,1),
                                                  datetime_end=datetime(2022,12,31))\
                         .strftime("%Y/%m/%d, %H:%M:%S") for i in range(p)]
df['wallet_price'] = [faker.pricetag() for i in range(p)]
dimwallet = df.copy()
dimwallet
```

```python
p = 10
df = pd.DataFrame()
df['customer_id'] = [faker.ssn() for i in range(p)]
df['first_name'] = [faker.first_name() for i in range(p)]
df['last_name'] = [faker.last_name() for i in range(p)]
df['gender'] = ccas_cat(p, ['m','f'], ratios=[38,62])
df['birth_date'] = [faker.date_of_birth() for i in range(p)]
df['email'] = df.apply(lambda row: "{}.{}@gmail.com".format(row.first_name.lower(),
                                                            row.last_name.lower()), axis=1)
df['city'] = [faker.city() for i in range(p)]
df['phone'] = [faker.phone_number() for i in range(p)]
df['start_date'] = [faker.date_time_between_dates(datetime_start=datetime(2021,1,1),
                                                  datetime_end=datetime(2022,3,30))\
                         .strftime("%Y/%m/%d, %H:%M:%S") for i in range(p)]
df['end_date'] = [faker.date_time_between_dates(datetime_start=datetime(2022,4,1),
                                                  datetime_end=datetime(2022,9,30))\
                         .strftime("%Y/%m/%d, %H:%M:%S") for i in range(p)]
dimcustomer = df.copy()
dimcustomer
```

```python
p = 100
df = pd.DataFrame()
df['transaction_id'] = np.arange(0, p, 1).tolist()
df['customer_id'] = ccas_cat(p, dimcustomer.customer_id.tolist())
df['transaction_date'] = [faker.date_time_between_dates(datetime_start=datetime(2022,1,1),
                                                       datetime_end=datetime(2022,9,30))\
                         .strftime("%Y/%m/%d, %H:%M:%S") for i in range(p)]
df['wallet_id'] = ccas_cat(p, dimwallet.wallet_id.tolist())
df['type'] = ccas_cat(p, ['credit','debit'], ratios=[65,35])
df['amount'] = [faker.pricetag() for i in range(p)]

transactions = df.copy()
transactions.head()
```

```python
dimcustomer.to_csv("customer.csv", index=False)
dimwallet.to_csv("wallet.csv", index=False)
transactions.to_csv("transactions.csv", index=False)
```
