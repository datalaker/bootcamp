```python
import glob
from dotenv import load_dotenv
import os
import pandas as pd
from sqlalchemy import create_engine

load_dotenv()


engine = create_engine(f"postgresql://{os.getenv('DB_STRING')}")

engine.execute("""
                CREATE SCHEMA IF NOT EXISTS olist
                """)

filenames = [os.path.basename(x) for x in glob.glob('./data/*.csv')]
for filename in filenames:
    dataset_name = filename.split('_dataset')[0]
    dataset = pd.read_csv(filename)
    dataset.to_sql(
        name=dataset_name, con=engine, schema="olist", if_exists="replace"
    )
    print(f'SUCCEED: {filename} -> {dataset_name}')
```

```python
query = """SELECT
    table_schema
    , table_name
    FROM information_schema.tables
    WHERE
    (
    table_schema = 'olist'
    )
    ORDER BY table_schema, table_name;"""

pd.read_sql_query(query, engine)
```
